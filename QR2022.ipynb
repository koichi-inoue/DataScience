## データサイエンス 授業アンケート結果

```python
# pandasをインポート
import pandas as pd

# 回答データの読み込み
df = pd.read_csv('https://raw.githubusercontent.com/koichi-inoue/DataScience2022/main/QR2022.csv')
df
```

```python
# データフレームの基本情報
print('行数・列数：', df.shape)
print('項目名：', df.columns)
print('各列のデータ型：\n', df.dtypes)
```

```python
# 要約統計量を表示
df.describe()
```

```python
# 相関係数を表示
df.corr()
```

```python
# 専攻別の平均｜全ての数値項目
df_mean_1 = df.groupby('専攻').mean()
df_mean_1.loc[ ['情報', '地域'],:]
```

```python
# 男女別の平均｜全ての数値項目
df_mean_2 = df.groupby('性別').mean()
df_mean_2.loc[ ['男性', '女性'],:]
```
