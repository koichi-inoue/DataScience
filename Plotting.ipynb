# グラフ描画サンプル 

## ライブラリの読み込み

```python
# 数値計算、ビジュアライゼーションライブラリの読み込み
import numpy as np
import matplotlib.pyplot as plt
```

付記：plot は「複数の点の位置を結ぶかたちでグラフを生成する」ものです。「式をそのままグラフにする」ものではありません。

## 一次関数のグラフ

```python
# １次元のベクトル（配列データ）の生成
x = np.array( range(0,100) )

# ベクトルと数値の演算（ベクトルの要素ごとに計算）＞ 出力 y もベクトル
y = 2 * x + 5

# プロット
plt.plot( x, y, label='y = 2x + 5' )
plt.legend()

# プロットの表示
plt.show()
```

## ２次関数のグラフ

```python
# データ生成　np.arange(x軸の最小値, x軸の最大値, 刻み)
x = np.arange(-4.0, 4.0, 0.1)
y = x**2 

# プロット
plt.plot( x, y, label='y = x^2' )
plt.legend()

# プロットの表示
plt.show()
```

## サインカーブ

```python
# データ生成　np.arange(x軸の最小値, x軸の最大値, 刻み)
x = np.arange( -2*np.pi, 2*np.pi, 0.1 )
y = np.sin(x)

# プロット
plt.plot( x, y, label = 'y = sin(x)' )
plt.legend()

# プロットの表示
plt.show()
```

## 複数のグラフを描く
* plot は、あらかじめpyplot に用意されたグラフの描画領域に具体的なグラフを描画します。
* 複数のグラフを描画する場合は、X軸かY軸か、いずれかの値を共通にするのが前提です。

```python
# データ生成
x = np.arange( -2*np.pi, 2*np.pi, 0.1 )

# グラフ用のデータを生成
y1 = np.sin(x)
y2 = np.cos(x)

# プロット
plt.plot( x, y1, label = 'y = sin(x)'  )
plt.plot( x, y2, label = 'y = cos(x)'  )
plt.legend()

# 表示
plt.show()


```
