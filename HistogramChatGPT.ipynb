以下の仕様に従って、Pythonのコードを書いて下さい。
1. 1000人分の国語、数学、英語の得点が入った sample.csv を読み込む
2. 国語の点数をヒストグラム化
3. 2で作成したヒストグラムに、正規分布曲線を重ねて表示

### OUT（ChatGPT の出力）

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# 1. CSVファイルからデータを読み込む
df = pd.read_csv('https://raw.githubusercontent.com/koichi-inoue/DataScience/main/Dummy500.csv')

# 国語の得点列を取得
kokugo_scores = df['国語']

# 2. 国語の点数をヒストグラム化
plt.hist(kokugo_scores, bins=20, density=True, alpha=0.6, color='b', label='国語得点ヒストグラム')

# 3. 正規分布曲線を重ねて表示
mu, std = np.mean(kokugo_scores), np.std(kokugo_scores)
x = np.linspace(min(kokugo_scores), max(kokugo_scores), 100)
p = norm.pdf(x, mu, std)

plt.plot(x, p, 'k', linewidth=2, label='正規分布曲線')

plt.xlabel('得点')
plt.ylabel('確率密度')
plt.legend()
plt.title('国語得点のヒストグラムと正規分布')
plt.grid()
plt.show()


```

* データの所在のみ実態に合わせて書き換えていますが、それ以外は ChatGPTの出力そのままです。実行結果を見ても正しくコーディングされていることがわかります。
* 要求仕様に記載した 1000件というデータ数は、プログラムには不要です（データを読み込んだ時点で自動的にカウントされます）。その点はちゃんと無視されています。
